```python
import requests
from bs4 import BeautifulSoup
```

```python
# Make a get-request
pref = "https://archive.ics.uci.edu/ml/"
url = "https://archive.ics.uci.edu/ml/datasets.php"
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
datasets = set()
for link in soup.find_all('a'):
    if link.get("href").split("/")[0] == "datasets":
        datasets.add(pref+link.get("href"))
```

## Collect all datasets links 

```python
for link in datasets:
    print(link)
```
